```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1.2

xm = np.linspace(-1,1,100)
ym = np.linspace(-1,1,100)
xm,ym = np.meshgrid(xm,ym)

zm = xm**2./2. + ym**2. * a/2.

plt.contour(xm,ym,zm,100)
plt.axis('square')
```

```python
L = max(a,1)
s = 2./L
#s = 1.5/L
def get_grad(x) : return np.array([x[0],x[1]*a])

    
```

```python


plt.figure()
plt.contour(xm,ym,zm,100)
plt.axis('square')
x = np.array([1,0])
plt.plot(x[0],x[1],'k.',markersize=10.)
for iter in xrange(100):
    x = x - s*get_grad(x)
    plt.plot(x[0],x[1],'k.',markersize=10./np.sqrt(iter+2.))
```
