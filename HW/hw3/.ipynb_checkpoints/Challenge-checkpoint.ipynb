```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.io as sio

from collections import Counter
import random

from tabulate import tabulate
```

```python
def plotMyHist(p, mu, sigma, m=1000):
    
    elements = [1,-1]
    probabilities = [p, 1-p]    
    m =100000
    Y = np.random.choice(elements, m, p=probabilities)   
    X = np.random.normal(Y * mu, sigma, m)
    count, bins, ignored = plt.hist(X, 300, density=True)
    plt.title("Gaussian Distribution of X = N(" + "Y*"+str(mu)+ ","+ str(sigma) +") with p = " + str(p))
    plt.show()
    return
```

## 1a)

```python
pValues = [0.1, 0.25, 0.5]
mu = 2
sigma = 1
for p in pValues:
    plotMyHist(p, mu, sigma)
```

## 1b)

```python
muValues = [0.1,2,10]
p = 0.5
sigma = 1
for mu in muValues:
    plotMyHist(p, mu, sigma)
```

## 1c)

```python
sigmaValues = [0.1,1,3]
p = 0.5
mu = 2
for sigma in sigmaValues:
    plotMyHist(p, mu, sigma)
```
